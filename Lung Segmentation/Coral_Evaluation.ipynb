```python
import os
import numpy as np
from skimage import io

def convert_images_to_numpy(image_path, aug_img_path):
    # Get the list of all image files
    image_files = [f for f in os.listdir(image_path) if f.endswith('.png')]
    aug_image_files = [f for f in os.listdir(aug_img_path) if f.endswith('.png')]
    
    # Create the image_numpy folder if it doesn't exist
    if not os.path.exists('image_numpy'):
        os.makedirs('image_numpy')
        
    # Loop through each image file
    for i, image_file in enumerate(image_files):
        # Load the image file
        image = io.imread(os.path.join(image_path, image_file))
        
        # Convert the image to a numpy array
        image = np.array(image)
        
        # Save the numpy array as a .npy file
        np.save(os.path.join('image_numpy', 'image_{}.npy'.format(i)), image)
        
    # Loop through each augmented image file
    for i, aug_image_file in enumerate(aug_image_files):
        # Load the augmented image file
        aug_image = io.imread(os.path.join(aug_img_path, aug_image_file))
        
        # Convert the augmented image to a numpy array
        aug_image = np.array(aug_image)
        
        # Save the numpy array as a .npy file
        np.save(os.path.join('image_numpy', 'aug_image_{}.npy'.format(i)), aug_image)

# Call the function
convert_images_to_numpy(image_path, aug_img_path)
```

```python
import numpy as np
import os
from sklearn.model_selection import train_test_split

# set the path to the data folder
data_folder = "image_numpy"

# load the data from the data folder
data = []
filenames = []
for filename in os.listdir(data_folder):
    if filename.endswith(".npy"):
        filepath = os.path.join(data_folder, filename)
        data.append(np.load(filepath))
        filenames.append(filename)
data = np.array(data)

# split the data into training, testing, and validation sets
train_data, testval_data, train_filenames, testval_filenames = train_test_split(data, filenames, test_size=0.2,random_state = 2018)
test_data, val_data, test_filenames, val_filenames = train_test_split(testval_data, testval_filenames, test_size=0.5,random_state = 2018)

# create three folders to save the training, testing, and validation data
train_folder = "train_image_numpy"
test_folder = "test_image_numpy"
val_folder = "val_image_numpy"
if not os.path.exists(train_folder):
    os.makedirs(train_folder)
if not os.path.exists(test_folder):
    os.makedirs(test_folder)
if not os.path.exists(val_folder):
    os.makedirs(val_folder)

# save the training data into the train_folder with the original filenames
for i in range(train_data.shape[0]):
    filename = os.path.join(train_folder, train_filenames[i])
    np.save(filename, train_data[i])

# save the testing data into the test_folder with the original filenames
for i in range(test_data.shape[0]):
    filename = os.path.join(test_folder, test_filenames[i])
    np.save(filename, test_data[i])

# save the validation data into the val_folder with the original filenames
for i in range(val_data.shape[0]):
    filename = os.path.join(val_folder, val_filenames[i])
    np.save(filename, val_data[i])
```

```python
import os

folder1 = "image_numpy"
folder2 = "mask_numpy"

# get a list of filenames in each folder
folder1_filenames = set(os.listdir(folder1))
folder2_filenames = set(os.listdir(folder2))

# compute the set intersection to get the common filenames
common_filenames = folder1_filenames.intersection(folder2_filenames)

# remove files that are not in both folders
for filename in folder1_filenames:
    if filename not in common_filenames:
        os.remove(os.path.join(folder1, filename))

for filename in folder2_filenames:
    if filename not in common_filenames:
        os.remove(os.path.join(folder2, filename))
```

```python
import os

folder = "mask_numpy"
word_to_remove = "_mask"

for filename in os.listdir(folder):
    # check if the filename contains the word to remove
    if word_to_remove in filename:
        # remove the word from the filename
        new_filename = filename.replace(word_to_remove, "")
        # construct the full paths to the old and new files
        old_path = os.path.join(folder, filename)
        new_path = os.path.join(folder, new_filename)
        # rename the file
        os.rename(old_path, new_path)
```
